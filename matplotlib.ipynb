# Como usar la libreria de matplotlib 

## Linea simple

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y) # dibuja la linea

# titulos 
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico de Línea")

plt.show() # muestra la grafica 
```

## Grafica de dispercion 

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', marker='o') # crea la grafica de puntos
plt.title("Gráfico de Dispersión") # titulo
plt.show() # muestra la grafica
```

## Histograma 

```python
data = np.random.randn(1000) 


plt.hist(data, bins=100, color='blue', alpha=0.7) # crea el histograma
# bins: cantidad de barras
# alpha: transparencia

plt.title("Histograma") # titulo
plt.show() # muestra la grafica
```

## Grafico de barras

```python
categorias = ['A', 'B', 'C', 'D']
valores = [5, 7, 3, 8]

plt.bar(categorias, valores, color=['red', 'blue', 'green', 'purple']) # crea el diagrama de barras verticales 
# color: los colores de las barras

plt.title("Gráfico de Barras") # titulo
plt.show() # muestra la grafica
```
